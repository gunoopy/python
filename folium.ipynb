## Data
> latitude, longitude 정보

```python
np.random.seed(42)
N = 20

data = pd.DataFrame({
    'lat' : np.random.uniform(low=35, high=38, size=N), # latitude
    'long' : np.random.uniform(low=127, high=129, size=N) # longitude
})

popups = [str(i) for i in range(N)] # text info

data.head()
```

<br/>
<br/>
<br/>
<br/>

## Marker

```python
center = [36, 128]

m = folium.Map(center, zoom_start=7)

for i in range(N) :
    lat, long = data.iloc[i, :]
    title = popups[i]
    
    folium.Marker([lat, long], tooltip = title).add_to(m)
    
m
```

## Marker Cluster

```python
from folium import plugins

center = [36, 128]

m = folium.Map(center, zoom_start=7)

plugins.MarkerCluster(data, popups=popups).add_to(m)

m
```

## Circle Marker

```python
center = [36, 128]

m = folium.Map(center, zoom_start=7)

for i in range(N) :
    lat, long = data.iloc[i, :]
    title = popups[i]
    
    folium.CircleMarker([lat, long], color='red',radius = 20, tooltip=title).add_to(m)
    
m
```

## Poly Line

```python
path = data[:5]
path
```

<br/>
<br/>

```python
center = [36, 128]

m = folium.Map(center, zoom_start=7)

for i in range(N) :
    lat, long = data.iloc[i, :]
    title = popups[i]
    
    folium.CircleMarker([lat, long], color='red',radius = 10, tooltip=title).add_to(m)
    
folium.PolyLine(
    locations = path,
    tooltip = 'PolyLine'
).add_to(m)

m
```

## Rectangle

```python
center = [36, 128]

m = folium.Map(center, zoom_start=7)

for i in range(N) :
    lat, long = data.iloc[i, :]
    title = popups[i]
    
    folium.CircleMarker([lat, long], color='red',radius = 10, tooltip=title).add_to(m)

folium.Rectangle(
    bounds = path,
    tooltip = 'Rectangle'
).add_to(m)

m
```

## Polygon

```python
center = [36, 128]

m = folium.Map(center, zoom_start=7)

for i in range(N) :
    lat, long = data.iloc[i, :]
    title = popups[i]
    
    folium.CircleMarker([lat, long], color='red',radius = 10, tooltip=title).add_to(m)

folium.Polygon(
    locations = path,
    fill = True,
    tooltip = 'Polygon'
).add_to(m)

m
```

## Polygon with Convex Hull
> Convex Hull : 볼록 다각형

```python
from scipy.spatial import ConvexHull

ch = ConvexHull(data)

print(ch.vertices) # Convex Hull index
```

```python
data.iloc[ch.vertices, :] # Convex Hull 좌표
```

<br/>
<br/>

```python
center = [36, 128]

m = folium.Map(center, zoom_start=7)

for i in range(N) :
    lat, long = data.iloc[i, :]
    title = popups[i]
    
    folium.CircleMarker([lat, long], color='red',radius = 10, tooltip=title).add_to(m)

folium.Polygon(
    locations = data.iloc[ch.vertices, :],
    fill = True,
    tooltip = 'Polygon'
).add_to(m)

m
```
