# NetworkX
- Undirected Graph : 방향이 없는 그래프
- Directed Graph : 방향이 있는 그래프

```python
import networkx as nx
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10, 6) # 그래프 크기
```

## Undirected Graph

```python
# 그래프 생성
G = nx.Graph()

# 노드 추가
G.add_node('A')
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)

# G.add_nodes_from(['A', 1, 2, 3, 4]) # 한번에 추가하기

# 노드 조회
G.nodes
```

```python
# 그래프 그리기
nx.draw(G, with_labels = True)
plt.show()
```

```python
# 엣지 추가
G.add_edge(1, 'A')
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge('A', 3)
G.add_edge('A', 2)

# G.add_edges_from([(1, 'A'), (1, 2), (2, 3), ('A', 3), ('A', 2)]) # 한번에 추가하기

# 엣지 조회
G.edges
```

```python
# 그래프 그리기
nx.draw(G, with_labels = True)
plt.show()
```

```python
# 위치 고정
pos = nx.spring_layout(G)
```

```python
# 그래프 그리기 (여러 번 실행)
nx.draw(G, pos = pos, with_labels = True)
plt.show()
```

```python
# 엣지 추가(weight)
G.add_edge('A', 2, weight = 3)
G.edges
```

```python
# weight 정보 추출
labels = nx.get_edge_attributes(G, 'weight')
labels
```

```python
# 그래프 그리기
nx.draw(G, pos = pos, with_labels = True)
nx.draw_networkx_edge_labels(G, pos = pos, edge_labels = labels)
plt.show()
```

## Directed Graph

```python
# 객체 생성
G = nx.DiGraph()

# 노드 추가
G.add_node('A')
G.add_node(1)
G.add_node(2)
G.add_node(3)

G.nodes
```

```python
# 엣지 추가
G.add_edge(1, 'A', weight = 3)
G.add_edge(1, 2, weight = 3)
G.add_edge(2, 3, weight = 3)
G.add_edge('A', 3, weight = 3)
G.add_edge('A', 2, weight = 3)

G.edges
```

```python
# 위치 고정, weight 정보 추출
pos = nx.spring_layout(G)
labels = nx.get_edge_attributes(G, 'weight')
```

```python
# 그래프 그리기
nx.draw(G, pos = pos, with_labels = True)
nx.draw_networkx_edge_labels(G, pos = pos, edge_labels = labels)
plt.show()
```

## 데이터프레임 적용

```python
import pandas as pd

df = pd.DataFrame({
    'from' : list('ABCAED'),
    'to' : list('DAECAE'),
    'weight' : [1, 2, 3, 4, 5, 6]
})
df
```

```python
# 데이터프레임으로부터 그래프 생성
G = nx.from_pandas_edgelist(df, 'from', 'to', 'weight')
```

```python
# 위치 고정, weight 정보 추출
pos = nx.spring_layout(G)
labels = nx.get_edge_attributes(G, 'weight')
```

```python
# 그래프 그리기
nx.draw(G, pos = pos, with_labels = True)
nx.draw_networkx_edge_labels(G, pos = pos, edge_labels = labels)
plt.show()
```

<br/>
<br/>
<br/>
<br/>

## 기타 옵션

```python
nx.draw(G,
        pos = pos, # 위치
        with_labels = True, # 라벨 표시
        font_color = 'w', # 폰트 흰색
        font_weight = 'bold', # 볼드체
        font_size = 30, # 폰트 크기
        node_size = 3000, # 노드 크기
        node_color = range(G.number_of_nodes()), # 노드 색깔
        cmap = plt.cm.Reds) # color map
nx.draw_networkx_edge_labels(G,
                             pos = pos, # 위치
                             edge_labels = labels, # 표시할 라벨
                             font_size = 20, # 폰트 크기
                             bbox = {'facecolor':'y','edgecolor':'r','boxstyle':'round','alpha':0.5}, # 바운딩 박스
                             rotate = False) # 회전 여부
plt.show()
```

## 연습문제

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import itertools

plt.rcParams['figure.figsize'] = (16, 8) # 그래프 크기
```

```python
# itertools.combinations(리스트, n) : 리스트의 값들을 n개씩 묶음
list(itertools.combinations(['감자칩', '맥주', '쥐포', '빵'], 2))
```

```python
# 구매 목록
purchases = [['감자칩', '맥주', '쥐포', '빵'],
             ['감자칩', '맥주', '쥐포'],
             ['감자칩', '맥주'],
             ['감자칩', '콜라'],
             ['기저귀', '맥주', '쥐포', '빵'],
             ['기저귀', '맥주', '쥐포'],
             ['기저귀', '맥주'],
             ['기저귀', '맥주']]

# 각 구매 목록에서 2개씩 짝을 지어 데이터 프레임에 저장하세요.

df = pd.DataFrame() # 빈 데이터 프레임

for purchase in purchases :
    for s, t in itertools.combinations(purchase, 2) :
        df = df.append([[s, t, 1]])
        
df.columns = ['source', 'target', 'weight']
df
```

```python
# 'source', 'target' 별 개수를 구하세요.
df = df.groupby(['source', 'target']).sum()
df
```

```python
# 인덱스를 초기화하세요.
df = df.reset_index()
df
```

```python
# 'weight' 열을 기준으로 내림차순으로 정렬하세요.
df = df.sort_values('weight', ascending = False)
df
```

```python
# 그래프를 생성하세요.
G = nx.from_pandas_edgelist(df, 'source', 'target', 'weight')
```

```python
# 노드를 조회하세요.
G.nodes
```

```python
# edge를 조회하세요.
G.edges
```

```python
# 위치를 고정하고 weight 정보를 추출하세요.
pos = nx.spring_layout(G)
labels = nx.get_edge_attributes(G, 'weight')
labels
```

```python
# 그래프를 그리세요.
nx.draw(G,
        pos = pos,
        with_labels = True,
        font_color = 'w',
        font_weight = 'bold',
        font_family = 'Malgun Gothic',
        font_size = 20,
        node_size = 5000,
        node_color = 'c')
nx.draw_networkx_edge_labels(G,
                             pos = pos,
                             edge_labels = labels,
                             font_size = 20,
                             rotate = False)
plt.show()
```

```python
# 노드를 조회하세요.
G.nodes
```

```python
# weight가 4로 연결된 노드들의 색깔을 red로 설정하세요.
colors = ['red', 'red', 'red', 'c', 'c', 'c']
colors
```

```python
# 노드의 색깔을 colors로 지정하여 그래프를 그리세요.
nx.draw(G,
        pos = pos,
        with_labels = True,
        font_color = 'w',
        font_weight = 'bold',
        font_family = 'Malgun Gothic',
        font_size = 20,
        node_size = 5000,
        node_color = colors)
nx.draw_networkx_edge_labels(G,
                             pos = pos,
                             edge_labels = labels,
                             font_size = 20,
                             rotate = False)
plt.show()
```

```python
# weight가 4 이상으로 연결된 노드를 선택하세요.
main_nodes = []
for (x, y), weight in labels.items() :
    if weight >= 4 :
        main_nodes.append(x)
        main_nodes.append(y)
        
main_nodes = list(set(main_nodes))
main_nodes
```

```python
# main_nodes에 해당되면 'red', 그렇지 않으면 'c' 으로 색깔을 지정하세요.
colors = ['red' if node in main_nodes else 'c' for node in G.nodes]
colors
```

```python
# 노드의 색깔을 colors로 지정하여 그래프를 그리세요.
nx.draw(G,
        pos = pos,
        with_labels = True,
        font_color = 'w',
        font_weight = 'bold',
        font_family = 'Malgun Gothic',
        font_size = 20,
        node_size = 5000,
        node_color = colors)
nx.draw_networkx_edge_labels(G,
                             pos = pos,
                             edge_labels = labels,
                             font_size = 20,
                             rotate = False)
plt.show()
```
