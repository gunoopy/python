# How to set <span style = "color : red">c</span><span style = "color : orange">o</span><span style = "color : green">l</span><span style = "color : blue">o</span><span style = "color : purple">r</span> or <span style = "color : gray">*style*</span> according to some column
- seaborn scatterplot
- hue : 색 구분
    - palette : 색 지정
    - hue_order : 색 순서
- style : 스타일 구분
    - markers : 스타일 지정
    - style_order : 스타일 순서

<br/>
<br/>

## Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (8, 6)
```

```python
np.random.seed(42)

df = pd.DataFrame(np.random.randn(100, 2) + 3, columns = ['X', 'Y'])
df['Z'] = df['X'].apply(lambda x : 1 if x < 4 else -1)

display(df)
```

```python
sns.scatterplot(data = df, x = 'X', y = 'Y', s = 150) # s : size
plt.show()
```

<br/>
<br/>
<br/>
<br/>

## Set Color

- hue

```python
sns.scatterplot(data = df, x = 'X', y = 'Y', s = 150,
                hue = 'Z')
plt.show()
```

- palette

```python
sns.scatterplot(data = df, x = 'X', y = 'Y', s = 150,
                hue = 'Z', palette = 'pastel') # seaborn built-in palette
plt.show()
```

```python
sns.scatterplot(data = df, x = 'X', y = 'Y', s = 150,
                hue = 'Z', palette = ['red', 'blue']) # user palette
plt.show()
```

```python
sns.scatterplot(data = df, x = 'X', y = 'Y', s = 150,
                hue = 'Z', palette = {-1 : 'blue', 1 : 'red'})
plt.show()
```

- hue_order

```python
sns.scatterplot(data = df, x = 'X', y = 'Y', s = 150,
                hue = 'Z', palette = ['red', 'blue'], hue_order = [1, -1])
plt.show()
```

<br/>
<br/>
<br/>
<br/>

## Set Style

- style

```python
sns.scatterplot(data = df, x = 'X', y = 'Y', s = 150,
                style = 'Z')
plt.show()
```

- markers

```python
sns.scatterplot(data = df, x = 'X', y = 'Y', s = 150,
                style = 'Z', markers = ['s', '^'])
plt.show()
```

```python
sns.scatterplot(data = df, x = 'X', y = 'Y', s = 150,
                style = 'Z', markers = {1 : 's', -1 : '^'})
plt.show()
```

- style_order

```python
sns.scatterplot(data = df, x = 'X', y = 'Y', s = 150,
                style = 'Z', markers = ['s', '^'], style_order = [1, -1])
plt.show()
```

<br/>
<br/>
<br/>
<br/>

## Summary

```python
sns.scatterplot(data = df, x = 'X', y = 'Y', s = 150,
                hue = 'Z', palette = ['red', 'blue'], hue_order = [1, -1], # Color
                style = 'Z', markers = ['s', '^'], style_order = [1, -1], # Style
                edgecolor = 'k')
plt.show()
```
